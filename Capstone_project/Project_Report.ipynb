## **Introduction:**
Customer churn, the rate at which customers stop using a service, is a critical challenge for telecom companies. Identifying customers likely to churn allows companies to implement proactive retention strategies, thereby reducing revenue loss and maintaining customer satisfaction. In this project, we aim to develop a predictive model to identify customers at risk of churn for a telecom company.

**Project Scope:**
- Importation and Loading: Importing relevant Libraries for the project and loading the dataset(s).
- Data Cleaning: Cleaning the dataset to handle missing values, and inconsistencies.
- Data Visualization: Exploring data patterns and relationships through visualizations.
- Feature Engineering: Creating new features and transforming existing ones to improve model performance.
- Data Imbalance: Observed imbalances in the data are rectified.
- Machine Learning: Training predictive models using historical data with known churn outcomes.
- Model Performance Evaluation: The perormance of the model is evaluated using different metrics.

**Project Description:**
The project involves building a predictive model using machine learning algorithms trained on historical customer data. We employed data cleaning techniques to preprocess the dataset, handle missing values, and ensure data quality. Exploratory data analysis (EDA) helped us understand the characteristics of churned and non-churned customers, guiding feature selection and engineering.

We then proceed with feature engineering to transform existing features to enhance the predictive power of the model. Random forests supervised machine learning algorithm was then used to train the preprocessed data to predict customer churn.

**Dataset:**
The dataset used for this project is sourced from Kaggle and contains information about telecom customers, including demographics, services subscribed, and churn status. The target variable is the churn status (churned or not churned), while features include customer demographics, usage patterns(services), and account information.
