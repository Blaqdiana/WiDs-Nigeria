```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reading the data from a CSV file
df = pd.read_csv('data.csv')

# Printing the first 5 rows of the DataFrame
print(df.head())

# Printing the shape of the DataFrame - number of rows, number of columns
print(df.shape)

# Printing the number of missing values in each column
print(df.isna().sum())
```

```python
# Dropping the column named 'Unnamed: 32' from the DataFrame
df = df.drop('Unnamed: 32', axis=1)

# Printing the summary of the DataFrame
df.info()
```

```python
# Replacing the 'M' and 'B' values in the 'diagnosis' column with '1' and '0'
df["diagnosis"].replace({"M":"1", "B":"0"}, inplace=True)

# Converting the 'diagnosis' column to integer type
df["diagnosis"] = df["diagnosis"].astype("int")

# Printing the first 5 rows of the 'diagnosis' column
df["diagnosis"].head()
```

```python
# Counting the number of occurrences of each unique value in the 'diagnosis' column
df["diagnosis"].value_counts()
```

```python
# Creating a countplot for the 'diagnosis' column in the DataFrame
ax = sns.countplot(data=df, x="diagnosis")

# Setting the title of the plot
plt.title("Diagnosis")

# Setting the labels for the x-axis
plt.xticks([0, 1], ["Benign", "Malignant"])

# Setting the label for the y-axis
plt.ylabel("Count")

# Displaying the plot
plt.show()
```
